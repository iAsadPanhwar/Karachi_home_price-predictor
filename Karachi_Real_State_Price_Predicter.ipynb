```python
def convert_to_millions(price):
    if 'Crore' in price:
        return float(price.split(' ')[0]) * 10
    elif 'Lakh' in price:
        return float(price.split(' ')[0]) / 100
    else:
        return float(price)
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key , subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqyard)
        st = np.std(subdf.price_per_sqyard)
        reduced_df = subdf[(subdf.price_per_sqyard > (m-st)) & (subdf.price_per_sqyard > (m+st))]
        df_out = pd.concat([df_out,reduced_df], ignore_index = True)
    return df_out
```

```python
X_train, X_test, y_train, y_test = train_test_split(x , y, test_size = 0.1)
```

```python
cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 0)
```

```python
cross_val_score(LinearRegression(), x, y, cv = cv)
```

```python
def find_best_model_using_gridsearchcv(X, y):
    algos = {
        'linear_regression': {
            'model': LinearRegression()
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config.get('params', {}), cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])


find_best_model_using_gridsearchcv(x, y)
```

```python

def predict_price(location, type, sq_yard, bathrooms, bedrooms):
    
    loc_index = np.where(x.columns == location)[0]

    if len(loc_index) == 0:
        
        print(f"Location '{location}' not found in the feature matrix.")
        return None

    
    features = np.zeros(len(x.columns))
    
    
    features[0] = sq_yard
    features[1] = bathrooms
    features[2] = bedrooms
    
    
    features[loc_index[0]] = 1

    
    return model.predict([features])[0]
```
